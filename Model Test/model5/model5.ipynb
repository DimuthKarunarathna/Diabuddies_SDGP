```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow import keras
from tensorflow.keras import layers
```

```python
data = pd.read_excel('dataCustom.xlsx')
X = data[['carbs', 'calories', 'fiber', 'fat', 'protein', 'patient_number', 'Before_BGL', 'Drinker','gender','age','BMI','Smoking','diabetes_duration']]
data = data.astype('float64')
Y = data['After_BGL']
print(data.dtypes)
print(X)
# data.to_excel('dataCustom.xlsx', index=False)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# Normalize the input features using a layer
normalizer = layers.Normalization()
normalizer.adapt(X_train)

# Build a linear regression model using Keras API
model = keras.Sequential([
  normalizer,
  layers.Dense(1)
])

# Compile the model with loss function and optimizer
model.compile(loss='mean_squared_error', optimizer=tf.optimizers.Adam(0.1))

# Train the model for 100 epochs
history = model.fit(X_train, y_train, epochs=100,validation_data=(X_test, y_test))
# Evaluate the model on the test set
test_loss = model.evaluate(X_test, y_test)
```

```python
x_input = [[24.83, 9.80, 0.81, 1.00, 2.66, 77,118.8,0,1,35,20.6,0,10]]
y_pred_test = model.predict(x_input)
print(y_pred_test)
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save the TFLite model
with open('linear_regression_model.tflite', 'wb') as f:
    f.write(tflite_model)
```
