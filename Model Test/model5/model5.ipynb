```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_excel('alldata.xlsx')
X = data[['carbs', 'calories', 'fiber', 'fat', 'protein', 'patient_number', 'Before_BGL', 'Drinker','Hour',]]
le = LabelEncoder()

data['Drinker'] = le.fit_transform(data['Drinker'])
data['patient_number'] = le.fit_transform(data['patient_number'])

print(data['Drinker'])
data.to_excel('dataCustom.xlsx', index=False)
```

```python
data = pd.read_excel('dataCustom.xlsx')
X = data[['carbs', 'calories', 'fiber', 'fat', 'protein', 'patient_number', 'Before_BGL', 'Drinker','Gender','Age','BMI','Smoking','Drinker','diabetes_duration']]
print(X)
Y = data['After_BGL']
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, Y, test_size = 0.2)
# Define the model
model = Sequential()
model.add(Dense(1, input_shape=(14,), activation='linear'))
# Compile the model
model.compile(optimizer='rmsprop', loss='mse', metrics = ['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))
y_pred = model.predict(X_test)
# mse = mean_squared_error(y_test, y_pred)
# r2 = r2_score(y_test, y_pred)
# print('MSE:', mse)
# print('R^2:', r2)
```
